# Array math

```python
import numpy as np
x = np.array([[1,2],[3,4]], dtype = np.float64)
y = np.array([[5,6],[7,8]])

print(x+y)  
print(np.add(x,y))
print()

print(x)
print()

print(x - y)
print(np.subtract(x,y))
print()

print(x*y)
print(np.multiply(x,y))
print()

print(x.dot(y))
print()

print(x/y)
print(np.divide(x,y))
print()

print(np.sqrt(x))
print()

v = np.array([10,12])
w = np.array([20,15])

print(v.dot(w))
print()
print(np.dot(v,w))
print()

#use @ at dot operator
print(v @ w)
print()

print(x @ y)
print()
```

```python
import nltk
from nltk.corpus import movie_reviews
nltk.download('movie_reviews')
movie_reviews.fileids()
movie_reviews.categories()

print("Positive movie reviews : ",len(movie_reviews.fileids('pos')))
print("Negative movie reviews : ",len(movie_reviews.fileids('neg')))

import matplotlib.pyplot as plt
fig = plt.figure(figsize=(5,5))
labels = 'Positive','Negative'
sizes = [len(movie_reviews.fileids('pos')),len(movie_reviews.fileids('neg'))]
plt.pie(sizes,labels=labels,autopct='%1.1f%%',shadow=True,startangle=90)
plt.axis('equal')
plt.show()


print( "\n----------------------------------------------------------------------------------------------------------")
#tokenize string

review = movie_reviews.raw('pos/cv944_13521.txt')
print('\033[92m'+review)
review_tokens = review.split()
print()
print('\033[94m'+'Tokenized string:')
print(review_tokens)
print("\n -----------------------------------------------------------------------------------------------------------")

# remove stopword
from nltk.corpus import stopwords
import string

nltk.download('stopwords')
stopwords_english = stopwords.words('english')
print('Stop words\n')
print(stopwords_english)
print('\nPunctuation\n')
print(string.punctuation)

print('\033[92m')
print(review_tokens)

review_clean = []

for word in review_tokens:
  if( word not in stopwords_english and word not in string.punctuation ):
    review_clean.append(word)

print('\033[94m')
print('removed stop words and puntuations:\n')
print(review_clean)
print("\n -----------------------------------------------------------------------------------------------------------")

# stemming 
from nltk.stem import PorterStemmer
print('\033[92m')
print(review_clean)

stemmer = PorterStemmer()

review_stem = []

for word in review_clean:
  stem_word = stemmer.stem(word)
  review_stem.append(stem_word)

print('\033[94m')
print('stemmed words:')
print(review_stem)
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
print('\033[92m')
print(review_clean)

wordnet_lemmatizer = WordNetLemmatizer()

review_lem = []

for word in review_clean:
  lem_words = wordnet_lemmatizer.lemmatize(word)
  review_lem.append(lem_words)

print('\033[94m')
print('lemmatized words:')
print(review_lem)
```
